# Perform Normalization using loop

```python
dfn.head()
```

```python
for (columnName, columnData) in dfn.iteritems():    #loop statement
    if(dfn[columnName].dtype == 'float64' and columnName != 'Profit'):
        dfn[[columnName]] = n.fit_transform(dfn[[columnName]])
dfn.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
stand= StandardScaler()
dfs.head()
```

```python
for (columnName, columnData) in dfs.iteritems():    
    if(dfs[columnName].dtype == 'float64' and columnName != 'Profit'):
        dfs[[columnName]] = stand.fit_transform(dfs[[columnName]])
dfs.head()
```

# Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
logt = FunctionTransformer(np.log1p)
```

```python
df3.head()
```

```python
for (columnName, columnData) in df3.iteritems():    
    if(df3[columnName].dtype == 'float64' and columnName != 'Profit'):
        df3[[columnName]] = logt.fit_transform(df3[[columnName]])
df3.head()
```
